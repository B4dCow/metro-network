```python
from metronetwork import Seoul
import metronetwork as mn
import matplotlib.pyplot as plt
import networkx as nx
from networkx.drawing.nx_pydot import graphviz_layout


seoul = Seoul()
edges = mn.load_edges()
stations = mn.load_stations()
G = seoul.make_seoul_metro_network()
pos = nx.kamada_kawai_layout(G)

plt.figure(figsize=(15, 8))


node_labels = {key: value[1] for key, value in stations.items()}
# add nodes
nx.draw_networkx_nodes(G, pos, node_size=50, node_color="#fcfcfc", alpha=0.5)
nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=10, font_family='Malgun Gothic')

# add edges
for key in edges:
    nx.draw_networkx_edges(
        G.to_undirected(),
        pos,
        edgelist=edges.get(key),
        width=2,
        edge_color=mn.metro_utils.line_color.get(key),
    )

plt.axis("off")
plt.show() 
```
